2. When was the highest number of tv/movies released over the years


```python
netflix_release_year = netflix_data.copy()
```

```python
netflix_release_year['release_year'].value_counts()
```

```python
# netflix_data['release_year'].value_counts().plot(kind = 'bar')
yearly_counts = netflix_release_year.groupby(['release_year', 'type']).size().unstack(fill_value=0)
print(yearly_counts)
```

```python
yearly_counts = yearly_counts.sort_index()

# Create a line plot
plt.figure(figsize=(15, 6))
plt.plot(yearly_counts.index, yearly_counts['Movie'], label='Movies', marker='o')
plt.plot(yearly_counts.index, yearly_counts['TV Show'], label='TV Shows', marker='o')
plt.xlabel('Year Released')
plt.ylabel('Count')
plt.title('Netflix TV Shows and Movies Released Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

3. NETFLIX CONTENT DATE ADDED

```python
netflix_date_added = netflix_data.copy()
```

```python
# Extracting YEAR (2016) from dataset with data_added as 'February 19, 2016'

# Convert 'date_added' to datetime objects
netflix_date_added['date_added'] = pd.to_datetime(netflix_date_added['date_added'], errors='coerce')

# Extract the year from the 'date_added' column
netflix_date_added['year_added'] = netflix_date_added['date_added'].dt.year

# Count the number of titles added in each year and categorize them by 'type'
yearly_added_counts = netflix_date_added.groupby(['year_added', 'type']).size().unstack(fill_value=0)


# Display the counts per year
print(yearly_added_counts)
```

```python
yearly_added_counts = yearly_added_counts.sort_index()

# Create a line plot
plt.figure(figsize=(15, 6))
plt.plot(yearly_added_counts.index, yearly_added_counts['Movie'], label='Movies', marker='o')
plt.plot(yearly_added_counts.index, yearly_added_counts['TV Show'], label='TV Shows', marker='o')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.title('Netflix TV Shows and Movies Added Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```
