Explore the countries with the highest total vaccination administered and visualise the daily_vaccinations_per_million of these countries

------

```python

# countries = covid19_vac.groupby('country')['total_vaccinations'].max().sort_values(ascending= False)[:10].index
countries = covid19_vac.groupby('country')['total_vaccinations'].max().sort_values(ascending= False)[:10]
countries
```

```python
# Filter the countries with the highest total_vaccinations records
countries = covid19_vac.groupby('country')['total_vaccinations'].max().sort_values(ascending= False)[:10]
# countries

# Reset index to make 'country' a regular column
countries_data = countries.reset_index()  

# Creating a new DataFrame with the same structure as covid19_vac
top_countries = pd.DataFrame(columns=covid19_vac.columns)

# Appending rows with the same 'country' and 'total_vaccinations' values
'''The code then iterates through each row in countries_data and extracts rows from covid19_vac 
with the same 'country' and 'total_vaccinations' values.'''
for index, row in countries_data.iterrows():
    country = row['country']
    total_vaccinations = row['total_vaccinations']
    
    # Extract rows from covid19_vac with the same 'country' and 'total_vaccinations'
    matching_rows = covid19_vac[(covid19_vac['country'] == country) & (covid19_vac['total_vaccinations'] == total_vaccinations)]
    
    # Append matching rows to the top_countries DataFrame
    top_countries = top_countries._append(matching_rows)

# Display the resulting DataFrame
top_countries
```

```python
'''Filtering the countries('China', 'India', 'United States', 'Brazil', 'Indonesia', 'Japan',
       'Bangladesh', 'Pakistan', 'Vietnam', 'Mexico') and plot a line graph  '''

filtered_data = covid19_vac[(covid19_vac["country"]=="China") | (covid19_vac["country"]=="India") | (covid19_vac["country"]=="United States")| (covid19_vac["country"]=="Brazil")| (covid19_vac["country"]=="Indonesia")| (covid19_vac["country"]=="Japan")| (covid19_vac["country"]=="Bangladesh")| (covid19_vac["country"]=="Pakistan")| (covid19_vac["country"]=="Vietnam")| (covid19_vac["country"]=="Mexico")]
filtered_data

plt.figure(figsize=(20,8))
# sns.lineplot(x=top_countries['date'], y=top_countries['total_vaccinations'], hue= top_countries['country'], errorbar=('ci', False))
sns.lineplot(data =filtered_data, x='date', y='daily_vaccinations_per_million', hue= 'country')
plt.title('Countries with the highest record of daily vaccinations')
plt.xlabel('Date')
plt.ylabel('Daily Vaccinations per Million');
```

Find out which country has the highest number of full vaccinated people
